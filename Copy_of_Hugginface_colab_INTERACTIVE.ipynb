```python
# Cell 1: Install libraries
!pip install -q transformers torch

# Cell 2: Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Cell 3: Interactive Model Downloader
import os
from transformers import AutoModel, AutoTokenizer
from google.colab import files

print("=" * 60)
print("   HuggingFace Model Downloader to Google Drive")
print("=" * 60)

# Popular models for reference
print("\n📚 Popular models you might want:")
print("  • bert-base-uncased")
print("  • gpt2")
print("  • distilbert-base-uncased")
print("  • roberta-base")
print("  • facebook/bart-large")
print("  • t5-small")
print("  • google/flan-t5-base")
print("  • microsoft/deberta-v3-base")
print("\n" + "=" * 60)

# Get model name from user
MODEL_NAME = input("\n🔍 Enter the HuggingFace model name: ").strip()

if not MODEL_NAME:
    print("❌ No model name provided. Please run again.")
else:
    # Configure save path
    BASE_PATH = "/content/drive/MyDrive/HuggingFace_Models"
    model_folder = os.path.join(BASE_PATH, MODEL_NAME.replace('/', '_'))

    # Create directory
    os.makedirs(model_folder, exist_ok=True)

    print(f"\n📥 Downloading: {MODEL_NAME}")
    print(f"💾 Save location: {model_folder}")
    print("\nThis may take a few minutes depending on model size...\n")

    try:
        # Download and save model
        print("⏳ Downloading model...")
        model = AutoModel.from_pretrained(MODEL_NAME)
        model.save_pretrained(model_folder)
        print("✓ Model saved")

        # Download and save tokenizer
        print("⏳ Downloading tokenizer...")
        tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
        tokenizer.save_pretrained(model_folder)
        print("✓ Tokenizer saved")

        print(f"\n✅ SUCCESS! Model saved to Google Drive")
        print(f"📂 Location: {model_folder}")

        # Ask if user wants to test
        test = input("\n🧪 Test loading the model? (yes/no): ").strip().lower()

        if test in ['yes', 'y']:
            print("\n⏳ Loading model from Google Drive...")
            loaded_model = AutoModel.from_pretrained(model_folder)
            loaded_tokenizer = AutoTokenizer.from_pretrained(model_folder)

            test_text = "Hello, this is a test!"
            inputs = loaded_tokenizer(test_text, return_tensors="pt")
            outputs = loaded_model(**inputs)

            print("✅ Model loaded and tested successfully!")
            print(f"Output shape: {outputs.last_hidden_state.shape}")

        # Ask if user wants to download another
        another = input("\n🔄 Download another model? (yes/no): ").strip().lower()
        if another in ['yes', 'y']:
            print("\n👉 Run this cell again to download another model!")

    except Exception as e:
        print(f"\n❌ ERROR: {str(e)}")
        print("\nTroubleshooting tips:")
        print("  • Check if the model name is correct")
        print("  • For gated models (like Llama), you need a HuggingFace token")
        print("  • Check your internet connection")

# Cell 4: List all downloaded models
def list_models():
    BASE_PATH = "/content/drive/MyDrive/HuggingFace_Models"
    if os.path.exists(BASE_PATH):
        models = [d for d in os.listdir(BASE_PATH)
                 if os.path.isdir(os.path.join(BASE_PATH, d))]
        if models:
            print("\n📦 Downloaded models in your Google Drive:")
            print("=" * 60)
            for i, model in enumerate(models, 1):
                model_path = os.path.join(BASE_PATH, model)
                # Get folder size
                size = sum(os.path.getsize(os.path.join(model_path, f))
                          for f in os.listdir(model_path)
                          if os.path.isfile(os.path.join(model_path, f)))
                size_mb = size / (1024 * 1024)
                print(f"{i}. {model} ({size_mb:.1f} MB)")
        else:
            print("No models downloaded yet.")
    else:
        print("No models directory found.")

list_models()
```
